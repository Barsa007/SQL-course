```python
 import pandas as pd
import numpy as np

# DataFrame with duplicates
books = pd.DataFrame({
    "title": ["1984", "Dune", "Dune", "Hamlet", "Hamlet", "Hamlet", "Emma"],
    "author": ["Orwell", "Herbert", "Herbert", "Shakespeare", "Shakespeare", "Shakespeare", "Austen"],
    "year": [1949, 1965, 1965, 1603, 1603, 1609, 1815]
})

print('1. Duplicates üîÅ')
print('.1 Use duplicated() to find duplicate rows in the books DataFrame ')
display(books[books.duplicated()])
print('\n')

print('.2 Show only the non-duplicated rows ')
display(books[~books.duplicated()])
print('\n')

print('.3 Use drop_duplicates() to keep only the last occurrence of each title ')
display(books.drop_duplicates(subset=["author"], keep="last"))
print('\n')

print('.4 How many duplicates exist in the author column?')
books[books.duplicated(subset=['author'])]




```

```python
# DataFrame with missing values
students = pd.DataFrame({
    "first_name": ["Liam", np.nan, "Noah", "Emma", "Olivia"],
    "last_name": ["Smith", np.nan, "Johnson", "Brown", "Wilson"],
    "age": [20, np.nan, 22, 19, 21],
    "major": ["Math", np.nan, "CS", "History", "Biology"],
    "gpa": [3.5, np.nan, np.nan, 3.7, 3.9],
    "credits": [30, np.nan, np.nan, 25, 40]
})

print('2. Missing Data üï≥Ô∏è')
print('.1 Count missing values per column in the students DataFrame ')
display(students.isna().sum().to_frame())
print()
print('.2 Add a column missing_count with the number of NaNs per row ')
students['number_of_missing'] = students.isna().sum(axis=1)
display(students)
print()
print('.3 Select only rows where gpa is NaN but first_name is not null ')
display(students[(students['gpa'].isna() & students['first_name'].notnull())])
print()
print('.4 Which column has the most missing values?')
display(students[students['number_of_missing'] == students['number_of_missing'].max()])
```

```python
students = pd.DataFrame({
    "first_name": ["Liam", np.nan, "Noah", "Emma", "Olivia"],
    "last_name": ["Smith", np.nan, "Johnson", "Brown", "Wilson"],
    "age": [20, np.nan, 22, 19, 21],
    "major": ["Math", np.nan, "CS", "History", "Biology"],
    "gpa": [3.5, np.nan, np.nan, 3.7, 3.9],
    "credits": [30, np.nan, np.nan, 25, 40]
})


print('4. Dropping NaN üöÆ')
print('1 Drop all rows that contain any NaN values ')
display(students.dropna())
print()
print('2 Drop only rows where all values are NaN ')
display(students.dropna(thresh=1) )
print()
print('3 Keep rows that have at least 3 non-missing values (thresh=3) ')
display(students.dropna(thresh=3) )
print()
print('4 Drop columns with more than 50% missing values')
display(students.dropna(thresh=4,axis=1) )
print()
```

```python
customers = pd.DataFrame({
    "cust_id": [1, 2, 3, 4],
    "name": ["Alice", "Bob", "Clara", "David"]
})

orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "cust_id": [1, 2, 2, 5],
    "amount": [250, 400, 150, 500]
})

print('5. Merging Tables üîó\n'
'Use the customers and orders DataFrames.\n')

print('1 Perform an inner join between customers and orders ')
display(pd.merge(customers, orders, on="cust_id", how="inner"))
print()

print('2 Perform a left join (all customers, matching orders if they exist) ')
display(pd.merge(customers, orders, on="cust_id", how="left"))
print()

print('3 Perform a right join (all orders, matching customers if they exist) ')
display(pd.merge(customers, orders, on="cust_id", how="right"))
print()

print('4 Perform an outer join with the _merge indicator ')
display(pd.merge(customers, orders, on="cust_id", how="outer", indicator=True))
print()

print('5 Create two small DataFrames and do a cross join')

df1 = pd.DataFrame({"student_id": [[1, 2, 3]]})
df2 = pd.DataFrame({"name": [["Alice", "Bob", "Charlie"]]})
display(df1.merge(df2, how="cross"))
```
